```python
from rst_articles.notebook import Article
art = Article()
```

```python
art.add_custom_words(
	"rST",
	"reST",
	"reStructuredText",
)
```

```python
art.set_config(
	project="Test Notebook",
	title="Test title",
	subtitle="Test subtitle",
	author="Test author",
	institution="Test institution",
	dark=False,
)
```

```python
art.set_abstract("""
This document serves as a test
case for document generation
using the reStructuredText (rST)
markup language. Its primary
purpose is to validate the
capabilities and accuracy of
the document processing
tool chain, encompassing parsing,
rendering, and output formatting.
The content within is
intentionally structured to
exercise various rST features,
including different heading
levels, text formatting
(e.g., emphasis, strong emphasis),
lists, block quotes, and
potentially more complex elements
like tables or directives,
depending on the specific scope
of the test. Ultimately, this
document's successful generation
will confirm the tool chain's
readiness to convert standard
reST source files into coherent,
well-formatted output documents
(such as HTML, PDF, or man pages).
""")
```

```python
art.set_bibliography("""
@article{Colab2025,
  title={Markup Languages},
  author={Colab, A.},
  journal={J. Demo},
  year={2025}
}

@article{FCite,
  title={Citation example},
  author={Colab, Asecond.},
  journal={J. Demo},
  year={2025}
}
""")
```

```python
art.set_definitions("""
.. new-def:: FCite
    :long: First Cite
    :desc: A definition to test the First Cite directives and roles
    :language: en

.. new-def:: rST
    :long: reStructuredText
    :search: reStructuredText
    :language: en
    :max_sents: 2
""")
```

```python
art.set_index("intro", "lorem")
```

```python
art.write("intro.rst", """
Introduction
############

V2

Bad first subpart
----------

A lightweight markup language
is great for revision control.

This is a citation about
markup [Colab2025]_ .

This is the frst citation of :abbrev:`FCite` :fcite:`FCite`

This is the second citation of :abbrev:`FCite` :fcite:`FCite`

This is a wikipedia defined :abbrev:`rST` (check it in the definitions at the end)

.. csv-table:: Document formats
   :header: "Format", "Difficulty", "Customization", "Auto-formatting"
   :widths: 20, 20, 20, 20

   "rST", 2, "High", "Yes"
   "LaTeX", 4, "High", "Yes"
   "HTML", 5, "Very High", "HaHa"
   "Word", 1, "Low", "No"

.. plot::
   :caption: figure 4. illustration for python
   
   import pandas as pd
   import matplotlib.pyplot as plt
   import io

   csv_string = '''
   Format,Difficulty,Customization,Auto-formatting
   rST,2,High,Yes
   LaTeX,4,High,Yes
   HTML,5,Very High,HaHa
   Word,1,Low,No
   '''.strip()

   df = pd.read_csv(io.StringIO(csv_string))
   df.sort_values(by='Difficulty', ascending=False, inplace=True)
   
   plt.bar(df['Format'], df['Difficulty'], color='skyblue', edgecolor='black')
   plt.xlabel('Format')
   plt.ylabel('Difficulty')
   plt.title('Difficulty of Document Formats')
   plt.tight_layout()
   plt.show()
""")
```
